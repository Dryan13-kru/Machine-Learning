```python
#memanggil library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#memanggil tabel
data = pd.read_csv("sales.csv")
data
```

```python
plt.scatter(data['W1'], data['W2'], s=50)
plt.show()
#membuat scatter plot dari dua kolom ('W1' dan 'W2') dalam DataFrame data, di mana sumbu x (horizontal) menunjukkan nilai dari 'W1' dan sumbu y (vertikal) menunjukkan nilai dari 'W2'.
```

```python
data = np.asarray(data[['W1','W2']])
print(data)
#array NumPy yang berisi data dari dua kolom tersebut. Hasilnya kemudian dicetak ke layar. Jadi, data sekarang menjadi array NumPy yang berisi nilai dari 'W1' dan 'W2' dalam setiap baris.
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.title('Dendrogram')
dendogram = sch.dendrogram(sch.linkage(data, method='ward'))
```

```python
plt.title('Dendogram')
dendrogram = sch.dendrogram(sch.linkage(data, method='average'))
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5)
output_euclidean = ac.fit_predict(data)

plt.scatter(data[:,0], data[:,1], c=output_euclidean, s=50, cmap='rainbow')
plt.show()
#melakukan pengelompokan hierarkis pada data 'W1' dan 'W2' ke dalam 5 kelompok dan menampilkan scatter plot dengan warna yang mencerminkan hasil pengelompokan tersebut.
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=5, affinity='manhattan', linkage='complete')
output_manhattan = ac.fit_predict(data)

plt.scatter(data[:,0], data[:,1], c=output_euclidean, s=50, cmap='rainbow')
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score
print(davies_bouldin_score(data, output_euclidean))
print(davies_bouldin_score(data, output_manhattan))
#Skor Davies-Bouldin memberikan indikasi sejauh mana suatu kelompok homogen dan terpisah satu sama lain. Semakin rendah skor Davies-Bouldin, semakin baik hasil pengelompokan, dengan nilai 0 menunjukkan pengelompokan yang ideal.
```
